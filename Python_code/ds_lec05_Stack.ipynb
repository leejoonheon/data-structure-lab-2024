```python
class Stack:
  def __init__(self):
    self.items = []

  def is_empty(self):
    return self.items == []

  def push(self, item):
    self.items.append(item)

  def pop(self):
    return self.items.pop()

  def peek(self):
    return self.items[len(self.items)-1]
  
  def size(self):
    return len(self.items)

  def pop_all(self):
    self.items.clear()
```

```python
s = Stack()
s.push("dog")
s.push(3.141592)
s.push(11111)
s.pop()
print(s.size())
```

```python
def reverse(str):
  st = Stack()
  for i in range(len(str)):
    st.push(str[i])

  out = ""
  while not st.is_empty():
    out += st.pop()

  return out
```

```python
input = "Test Seq 12345"
answer = reverse(input)

print(input)
print(answer)
```
